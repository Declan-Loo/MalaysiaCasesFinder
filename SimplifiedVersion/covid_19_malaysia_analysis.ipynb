Obtain a date from the user. Make sure to format the date string accordingly using validation.

```python
year = input("Enter a year: ")
month = int(input("Enter a month: "))
day = int(input("Enter a date: "))

if month < 10:
    month_string = "0" + str(month)
else:
    month_string = str(month)
    
if day < 10:
    day_string = "0" + str(day)
else:
    day_string = str(day)

#Format the date_string for the linear search upcoming.
date_string = ("{}-{}-{}".format(year,month_string,day_string))
```

Iterates through the covid_array and conduct a linear search to find whether the data matches the inputted date_string. If yes, output the date and the number of new Covid cases for that particular day. If there is no date found, output to the user that the date is not given

```python
found = False
max_daily_cases = 0
max_date = ""
for data in covid_array:
    if data[0] == date_string:
        print("For date {date_check}, there are {new_cases} new cases on that date.".format(date_check = data[0],new_cases = data[1]))
        found = True
    if int(data[1]) > max_daily_cases:
        max_daily_cases = int(data[1])
        max_date = data[0]
        
if found == False:
    print("Date given is not found in the archive. Please re-run the program again and enter a correct date.")
```

Find the maximum cases that Malaysia had and print out the number and the date that moment occured.

```python
print("The maximum daily cases Malaysia had for a day is {max_cases}, on date {maximum_date}.".format(max_cases = max_daily_cases,maximum_date = max_date))
```

### Compare which vaccine was widely used for boosters (between AZ, Pfizer, Sinovac)

```python
if ((overall_data['pfizer3'].sum()) > (overall_data['astra3'].sum()) and (overall_data['pfizer3'].sum()) > overall_data['sinovac3'].sum()):
    print("Pfizer is the most used vaccine shot for Malaysians citizens who took the booster")
elif ((overall_data['astra3'].sum()) > (overall_data['pfizer3'].sum())) and ((overall_data['astra3'].sum()) > (overall_data['sinovac3'].sum())):
    print("AstraZeneca is the most used vaccine shot for Malaysians citizens who took the booster")
elif (overall_data['sinovac3'].sum() > overall_data['astra3'].sum()) and (overall_data['sinovac3'].sum() > overall_data['pfizer3'].sum()):
    print("Sinovac is the most used vaccine shot for Malaysians citizens who took the booster")
```
